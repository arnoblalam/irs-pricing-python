```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
```

```python
def roll_measure(x):
    covariance = np.cov(x[:-1], x[1:])[0, 1]
    if covariance < 0:
        return(2 * np.sqrt(-covariance))
    else:
        return(0)
```

```python
# List of prices
prices = np.loadtxt("prices.txt")
print(f"Read in {prices.shape[0]} entries")
```

```python
# Compute price changes
price_changes = np.diff(prices)
print(f"Roll Measure: {roll_measure(price_changes)}")
```

```python
# Work in logs of prices
eps = 1E-15 # Add a little bit to the price to avoid log of zero
log_prices = np.log(prices + eps)
log_price_changes = np.diff(log_prices)
print(f"Roll Measure (log prices): {roll_measure(log_price_changes)}")
```

```python
plt.plot(prices) # Plot the prices
```

```python
# Plot the price changes
plt.plot(price_changes)
plt.title("Price Changes")
plt.ylim(-0.25,0.25)
```

```python
plt.hist(price_changes) # Histogram of the price changes
```

```python
np.round(np.mean(price_changes), 4)
```
